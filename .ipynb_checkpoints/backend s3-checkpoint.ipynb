```python
# Create the S3 object
try:
    obj = client.get_object(Bucket='moviesdb',Key='movies/genres.csv')
except Exception as e:
    print(e)

# Read data from the S3 object
data = pandas.read_csv(obj['Body'])
    
# Print the data frame
print('Printing the data frame...')
pandas.DataFrame(data=data)
```

```python
r = client.select_object_content(
        ExpressionType='SQL',
        Bucket='moviesdb',Key='movies/genres.csv',
        Expression="select * from s3object s where s.genre_id like '%16%'",
        InputSerialization = {'CSV': {"FileHeaderInfo": "Use"}},
        OutputSerialization = {'CSV': {}},
)
for event in r['Payload']:
    if 'Records' in event:
        records = event['Records']['Payload'].decode('utf-8')
        print(records)
```
